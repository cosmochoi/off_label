```python
def find_omop_matches(diag_list, omop_list):
    match = 0
    for icd in diag_list.keys():
        print(icd, ":")
        std_codes = diag_list[icd]
        for j in std_codes:
            if j in omop_list:
                print(j)
                #print("OMOP reported indication relationship for ", j)
                match += 1
    return match

def find_omop_matches2(diag_list, omop_list):
    match = 0
    for codes in diag_list:
        if codes in omop_list:
            print(codes)
            match += 1
    return match
def get_values(d):
    for v in d.values():
        if isinstance(v, dict):
            yield from get_values(v)
        else:
            yield v
            
def flatten_dict(nested_dict):
    values = list(get_values(nested_dict))
    return [item for sublist in values for item in sublist]
```
